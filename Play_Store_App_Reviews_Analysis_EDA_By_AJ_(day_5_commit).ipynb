**In 'size' column, there are two types of units, one is in KB and other is in MB so we have to convert from KB to MB and also there is another type text 'Varies with device' prersent in it.**

````python
# Function to convert KB to MB:

def convert_to_mb(val):
  try:
    if 'M' in val:
      return float(val[:-1])
    elif 'k' in val:
      return round(float(val[:-1])/1024, 4)
    else:
      return 0
  except:
    return val
````

````python
# convert KB to MB with using function
df1['Size'] = df1['Size'].apply(lambda x: convert_to_mb(x))
df1['Size'] = df1['Size'].apply(lambda x: float(x))

df1.head()
````

#### Chart - 16 - Understand the relationships between variables through Pair Plot

````python
# Pair Plot visualization code

Rating = df1['Rating']
Size = df1['Size']
Installs = df1['Installs']
Reviews = df1['Reviews']
Type = df1['Type']
Price = df1['Price']

p = sns.pairplot(pd.DataFrame(list(zip(Rating, Size, np.log(Installs), np.log10(Reviews), Price, Type)), columns=['Rating','Size', 'Installs', 'Reviews', 'Price','Type']), hue='Type')
p.fig.suptitle("Pairwise Plot - Rating, Size, Installs, Reviews, Price",x=0.5, y=1.0, fontsize=20)

plt.show()
````

Pair plot is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters. It also helps to form some simple classification models by drawing some simple lines or make linear separation in our data-set.
